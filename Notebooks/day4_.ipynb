```python
import cv2
import numpy as np
from deepface import DeepFace
from sklearn.metrics.pairwise import cosine_similarity
from google.colab.patches import cv2_imshow
from google.colab import files
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
from deepface import DeepFace
import numpy as np

# Extract embedding from your photo
result = DeepFace.represent(
    img_path=list(uploaded.keys())[0],
    model_name="Facenet",
    enforce_detection=False
)

embedding = result[0]["embedding"]

print("Type:", type(embedding))
print("Length:", len(embedding))
print("\nFirst 10 numbers:", embedding[:10])
print("\nMin value:", round(min(embedding), 4))
print("Max value:", round(max(embedding), 4))
```

```python
embedding_array = np.array(embedding)

print("Shape:", embedding_array.shape)
print("Data type:", embedding_array.dtype)
print("\nFull embedding:")
print(np.round(embedding_array, 3))
```

```python

print("Upload photo 1:")
upload1 = files.upload()
photo1 = list(upload1.keys())[0]

print("Upload photo 2:")
upload2 = files.upload()
photo2 = list(upload2.keys())[0]


emb1 = DeepFace.represent(
    img_path=photo1,
    model_name="Facenet",
    enforce_detection=False
)[0]["embedding"]

emb2 = DeepFace.represent(
    img_path=photo2,
    model_name="Facenet",
    enforce_detection=False
)[0]["embedding"]


emb1_array = np.array(emb1).reshape(1, -1)
emb2_array = np.array(emb2).reshape(1, -1)

score = cosine_similarity(emb1_array, emb2_array)[0][0]

print(f"\nPhoto 1 embedding length: {len(emb1)}")
print(f"Photo 2 embedding length: {len(emb2)}")
print(f"\nSimilarity score: {round(score * 100, 2)}%")

if score > 0.7:
    print("Result: Likely SAME person ")
else:
    print("Result: Likely DIFFERENT people ")
```
