```python
!pip install deepface -q

import numpy as np
from deepface import DeepFace
from sklearn.metrics.pairwise import cosine_similarity
from google.colab import files
import json

```

```python

print("Upload 3-4 photos of DIFFERENT people for the database:")
db_uploads = files.upload()

database = []

for filename, _ in db_uploads.items():
    embedding = DeepFace.represent(
        img_path=filename,
        model_name="Facenet",
        enforce_detection=False
    )[0]["embedding"]

    database.append({
        "name": filename.split(".")[0],
        "embedding": embedding
    })
    print(f"Added to database: {filename}")

print(f"\nDatabase has {len(database)} records")
```

```python
print("Upload the photo you want to SEARCH for:")
search_upload = files.upload()
search_filename = list(search_upload.keys())[0]


search_embedding = DeepFace.represent(
    img_path=search_filename,
    model_name="Facenet",
    enforce_detection=False
)[0]["embedding"]

search_array = np.array(search_embedding).reshape(1, -1)


results = []

for record in database:
    db_array = np.array(record["embedding"]).reshape(1, -1)
    score = cosine_similarity(search_array, db_array)[0][0]

    results.append({
        "name": record["name"],
        "score": round(score * 100, 2)
    })

results.sort(key=lambda x: x["score"], reverse=True)


print("\n=== TOP MATCHES ===\n")
for i, match in enumerate(results[:3]):
    if match["score"] > 85:
        label = "HIGH confidence"
    elif match["score"] > 70:
        label = "POSSIBLE match"
    else:
        label = "unlikely"

    print(f"#{i+1}  {match['name']}")
    print(f"     Score: {match['score']}%")
    print(f"     Status: {label}")
    print()
```
