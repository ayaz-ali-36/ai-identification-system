```python
import cv2
import numpy as np
import pandas as pd

from google.colab.patches import cv2_imshow


image = cv2.imread('download.jfif')

# What shape is this image?
print("Image shape:", image.shape)
print("This means:", image.shape[0], "rows (height)")
print("            ", image.shape[1], "columns (width)")
print("            ", image.shape[2], "color channels (BGR)")

# Total number of pixels
total_pixels = image.shape[0] * image.shape[1]
print("\nTotal pixels in this image:", total_pixels)
```

```python
# Look at the top-left corner pixel (row 0, column 0)
print("Top-left pixel value:", image[0, 0])
print("This pixel has:")
print("  Blue  channel:", image[0, 0, 0])
print("  Green channel:", image[0, 0, 1])
print("  Red   channel:", image[0, 0, 2])

# Look at a 5x5 block of pixels from top-left corner
print("\nFirst 5x5 block of pixels:")
print(image[0:5, 0:5])
```

```python
# Display the image inside Colab
cv2_imshow(image)
print("Image displayed above")
```

```python
# Convert to grayscale — only 1 number per pixel instead of 3
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

print("Original image shape:", image.shape)
print("Grayscale image shape:", gray.shape)
print("\nNotice: grayscale has no 3rd dimension")
print("Each pixel is now just 1 number (0=black, 255=white)")

# Show grayscale
cv2_imshow(gray)
```
